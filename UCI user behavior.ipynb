```python
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline as ImbPipeline
```

```python
X = intention.drop('Revenue', axis=1)
y = intention['Revenue'].astype(int)  # Convert target to integer

# Encoding categorical features
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), ['Administrative', 'Administrative_Duration', 'Informational', 'Informational_Duration', 
                                    'ProductRelated', 'ProductRelated_Duration', 'BounceRates', 'ExitRates', 'PageValues', 'SpecialDay']),
        ('cat', OneHotEncoder(), ['Month', 'OperatingSystems', 'Browser', 'Region', 'TrafficType', 'VisitorType', 'Weekend'])
    ])

# Define models
models = {
    'RandomForest': RandomForestClassifier(),
    'SVM': SVC(probability=True),
    'MLP': MLPClassifier(),
    'DecisionTree': DecisionTreeClassifier(),
    'XGBoost': XGBClassifier()
}

# Train and evaluate models
for name, model in models.items():
    print(f"Training {name}...")
    # Pipeline with oversampling
    pipeline = ImbPipeline(steps=[
        ('preprocessor', preprocessor),
        ('sampler', SMOTE()),  # Oversampling
        ('classifier', model)
    ])
    
    # Split data
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
    
    # Train model
    pipeline.fit(X_train, y_train)
    
    # Predict and evaluate
    y_pred = pipeline.predict(X_test)
    print(f"Classification Report for {name}:\n{classification_report(y_test, y_pred)}")
    print(f"ROC AUC Score for {name}: {roc_auc_score(y_test, pipeline.predict_proba(X_test)[:, 1])}")
    print(f"PR AUC Score for {name}: {average_precision_score(y_test, pipeline.predict_proba(X_test)[:, 1])}")
    print("\n")
```
